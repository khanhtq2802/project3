```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
```

```python
# Load the data into a pandas DataFrame
df = pd.read_csv("C:/Users/khanh/code/DATA_SCIENCE/data/not using images/train_set.csv")
df = df.values
```

```python
x = df[:, 347] #km
y = df[:,356] #price
plt.scatter(x, y, s=0.1)
plt.xlabel("km")
plt.ylabel("price")
plt.show
```

```python
x = df[:, 346] #year
y = df[:,356] #price
plt.scatter(x, y, s=1)
plt.xlabel("year")
plt.ylabel("price")
plt.show
```

```python
# Fit a linear regression model to the training data
a = 1.7

one = np.ones((x.shape[0], 1))
Xbar = np.concatenate((one, x), axis = 1)

reg = LinearRegression()
reg.fit(Xbar, y)
# Use the fitted model to predict the missing values in the 'km' column
y_pred = (reg.predict(x))

plt.scatter(x, 1/y_pred**a, s=1)
plt.xlabel('price')
plt.ylabel('km')
plt.show()
```
